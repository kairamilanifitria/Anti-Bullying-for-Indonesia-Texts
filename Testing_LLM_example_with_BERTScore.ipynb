```python
input = [
    "Ngapain kamu di sini, nggak ada gunanya!",
    "Bodoh banget sih pendapat kayak gitu!",
    "Kamu bikin semuanya hancur, dasar payah!",
    "Orang seperti kamu bikin malu aja!",
    "Komentar kamu sampah, nggak penting!"
]
```

```python
import pandas as pd

df = pd.DataFrame({'input': input, 'predictions': predictions, 'references': references})
df
```

```python
results = bertscore.compute(predictions=predictions, references=references, lang="ind")
results
```

```python
# calculate the average score on bertscore calculation

import numpy as np

avg_precision = np.mean(results['precision'])
avg_recall = np.mean(results['recall'])
avg_f1 = np.mean(results['f1'])

print(f"Average Precision: {avg_precision}")
print(f"Average Recall: {avg_recall}")
print(f"Average F1 Score: {avg_f1}")
```
